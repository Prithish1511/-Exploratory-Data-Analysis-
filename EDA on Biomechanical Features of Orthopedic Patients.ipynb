```python
#loading the dataset
data = pd.read_csv("column_2C_weka.csv")
print(plt.style.available)
plt.style.use('ggplot')
```

```python
data.columns

# 'pelvic_incidence' - the angle between the line perpendicular to the sacral endplate at its midpoint and a line connecting this point to the axis of the femoral head
# 'pelvic_tilt numeric' - the angle created by a line running from the sacral endplate midpoint to the center of the bifemoral heads and the vertical axis.
# 'lumbar_lordosis_angle' - the angle between the upper plane of the L1 lumbar vertebrae and the upper plane of the S1 sacral vertebrae.
#  'sacral_slope' - angle between a horizontal line and the orientation of the sacral plateau.
# 'pelvic_radius' - the distance from the hip axis to the posterior-superior corner of the S1 endplate.
# 'degree_spondylolisthesis' - Spondylolisthesis is the slippage of one vertebral body with respect to the adjacent vertebral body causing mechanical or radicular symptoms or pain.
```
